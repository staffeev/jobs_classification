```python
from navec import Navec
from sklearn.metrics.pairwise import cosine_similarity
```

```python
navec = Navec.load('../datasets/navec_hudlit_v1_12B_500K_300d_100q.tar')
```

```python
def calc_score(word_a, words_b):
    result = [(word_b, cosine_similarity([navec[word_a]], [navec[word_b]])[0, 0]) for word_b in words_b]
    result.sort(key=lambda e: e[1], reverse=True)
    return result
```

```python
calc_score('профессия', ['разнорабочий', 'широкий', 'профиль'])
```
