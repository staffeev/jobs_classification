```python
from navec import Navec
from sklearn.metrics.pairwise import cosine_similarity
```

```python
navec = Navec.load('../datasets/navec_hudlit_v1_12B_500K_300d_100q.tar')
```

```python
def calc_score(word_a, words_b):
    result = []
    for word_b in words_b:
        if word_b not in navec:
            continue
        result.append((word_b, cosine_similarity([navec[word_a]], [navec[word_b]])[0, 0]))
    result.sort(key=lambda e: e[1], reverse=True)
    return result
```

```python
calc_score('трудоустройство', ['механик', 'широкий', 'профиль', "работа", "должность", "профессия"])
```

```python
calc_score('занятость', ['разнорабочий', 'широкий', 'профиль', "работа", "должность", "профессия"])
```

```python
calc_score('должность', ['разнорабочий', 'широкий', 'профиль', "работа", "должность", "профессия"])
```

```python
def compare(word, ix):
    desc = df["normalized_description"].iloc[ix]
    print(*calc_score(word, desc.split()), sep="\n")
```

```python
compare("работа", 100)
```

```python
def dist_ix(df, ix_1, ix_2):
    vec1 = df.iloc[ix_1]["vector"]
    vec2 = df.iloc[ix_2]["vector"]
    return cosine_distances([vec1], [vec2])[0, 0]
```

```python
for i in range(len(df)):
    for j in range(len(df)):
        if i != j:
            print(dist_ix(df,i, j))
```
